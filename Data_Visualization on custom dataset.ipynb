```python

plt.figure(figsize=(7,5))
ax = sns.countplot(data=df,x="gender", hue="qualification",palette="RdBu")
plt.xticks(size=12)
plt.xlabel("Gender",size=14)
plt.xticks(size=12)
plt.ylabel("Count",size=12)
plt.title('Gender with respect to qualification', fontsize=20)
ax.set_xticklabels(ax.get_xticklabels(),rotation=40,ha='right')
total = (len(df))
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='center')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
ax = sns.countplot(data=df,x="gender", hue="covd_vaccine",palette="RdBu")
plt.xticks(size=12)
plt.xlabel("Gender",size=14)
plt.xticks(size=12)
plt.ylabel("Count",size=12)
plt.title('Gender with respect to Covid Vaccine', fontsize=20)
ax.set_xticklabels(ax.get_xticklabels(),rotation=40,ha='right')
total = (len(df))
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='center')
plt.tight_layout()
plt.show()
```

```python

# Removing the outliers in the data
import numpy as np
# IQR
Q1 = np.percentile(df['age'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df['age'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", df.shape)

# Upper bound
upper = np.where(df['age'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(df['age'] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)
 
print("New Shape: ", df.shape)
```

```python
upper = Q3 +1.5*IQR

lower = Q1- 1.5*IQR

print(upper)
print(lower)
```
