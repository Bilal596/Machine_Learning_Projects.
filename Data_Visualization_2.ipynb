```python
# import libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the dataset
tips=sns.load_dataset('tips')
tips.head()
```

```python
# scatter plot in matplotlib
plt.scatter(tips['total_bill'],tips['tip'])
# adding labels in the matplotlib graphs

plt.xlabel('Total_Bill')
plt.ylabel("Tips($)")
plt.title("Scatter plot of tips along with total bill paid")
```

```python
 #scatter plot in matplotlib
plt.scatter(tips['total_bill'],tips['tip'],marker="*",c=tips['size'])
# adding labels in the matplotlib graphs

plt.xlabel("Total_Tips($)")
plt.ylabel("Tip recieved($)")
plt.title("Scatter plot of tips with total bill paid")
plt.colorbar()
```

```python
 #scatter plot in matplotlib
plt.scatter(tips['day'],tips['tip'],marker="*",c=tips['size'])
# adding labels in the matplotlib graphs

plt.xlabel("Days")
plt.ylabel("Tip recieved($)")
plt.title("Scatter plot of tips on each day")
plt.colorbar()
```

# Interpretation of the above graph
- There is a direct relational between table size and total bill
- The table with the highest size gave the highest tip

```python
# Step-1 import libraries
from bokeh.plotting import figure, output_file,show
from bokeh .palettes import magma
import pandas as pd
import seaborn as sns

# Step-2 Load the datasets
df=sns.load_dataset('tips')

# Step-3 initiating the graphs
graph=figure(title='Bokeh Scatter Graphr')

# Step-4 Color Selection
color=magma(256)

# Step-5 Plotting the graph
graph.scatter(df['total_bill'],df['tip'],color=color)

# Step -6 Display the graph
show(graph)

# Step-7 Export the file

from bokeh.io import export_png
export_png(graph,filename="Scatter.png")
```

# Plotly graphs

```python
# make a scatter plot with plotly
import plotly.express as px

# dataset
tips=sns.load_dataset('tips')

# draw the graph

fig=px.scatter(tips,x='total_bill',y='tip',color='size',size='size',hover_data=(['day']))

# show the plot

fig.show()
```

```python
# make a scatter plot with plotly
import plotly.express as px

# dataset
tips=sns.load_dataset('tips')

# draw the graph

fig=px.scatter(tips,x='day',y='tip',color='total_bill',size='total_bill',hover_data=(['day']))

# show the plot

fig.show()
```

```python
# make a bar plot with plotly
import plotly.express as px

# dataset
tips=sns.load_dataset('tips')

# draw the graph

fig=px.bar(tips,x='day',y='tip',color='day',hover_data=(['day']))

# show the plot

fig.show()
```

```python
px.area(tips,x='day',y='tip',color='day',hover_data=(['day']))
```
